```python
# Importing libraries
from ultralytics import YOLO
from roboflow import Roboflow
import os
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Loading dataset from Roboflow
rf = Roboflow(api_key="VukmouBkOiMlkW8x0UrB")
project = rf.workspace("manoj-vqpbj").project("myproject-3jpq7-wczko")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
train_img_dir = os.path.join(dataset.location, "train", "images")
val_img_dir = os.path.join(dataset.location, "valid", "images")

train_images = os.listdir(train_img_dir)
val_images = os.listdir(val_img_dir)

print(f"Train set images : {len(train_images)}")
print(f"Validation images: {len(val_images)}")
```

```python
# Model setup and training
model = YOLO("yolov8x.pt")

model.train(
    data=os.path.join(dataset.location, "data.yaml"),
    epochs=22,
    imgsz=640,
    batch=16,
    name="yolov8x_OD",
    patience=15,
    workers=2,
    verbose=True
)
```

```python
# Model validation
metrics = model.val(data=os.path.join(dataset.location, "data.yaml"))
```

```python
# Overall metrics
mp = metrics.box.mp
mr = metrics.box.mr
map50_95 = metrics.box.map
map50 = metrics.box.map50

# Calculate mean F1 from mp and mr
mean_f1 = 2 * mp * mr / (mp + mr + 1e-6)

print("Overall mAP50-95:", map50_95)
print("Overall mAP50:", map50)
print("Overall Precision:", mp)
print("Overall Recall:", mr)
print("Overall F1 Score:", mean_f1)
```
